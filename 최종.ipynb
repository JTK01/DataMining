## I. 데이터 전처리

### 1. 필요 데이터 추출 및 가공

#### 1) 상주인구

```python
import pandas as pd

df1 = pd.read_csv("data\서울시 우리마을가게 상권분석서비스(상권-상주인구).csv", encoding="cp949")

pop1_22 = df1.loc[(df1.기준_년_코드 == 2022) & (df1.기준_분기_코드 == 4)].iloc[:,[4,5,8,9,10,11,12,13]]

pop1_22['ave_age1'] = (15*pop1_22.iloc[:,2]+25*pop1_22.iloc[:,3]+35*pop1_22.iloc[:,4]+45*pop1_22.iloc[:,5]+55*pop1_22.iloc[:,6]+65*pop1_22.iloc[:,7])/pop1_22.iloc[:,1]
pop1_22['sum_age1'] = 15*pop1_22.iloc[:,2]+25*pop1_22.iloc[:,3]+35*pop1_22.iloc[:,4]+45*pop1_22.iloc[:,5]+55*pop1_22.iloc[:,6]+65*pop1_22.iloc[:,7]

pop1_iloc = pop1_22.iloc[:,[0,1,-2,-1]]

pop1_iloc.rename(columns={'상권 코드': '상권_코드', '총 상주인구 수': '총_상주인구_수'}, inplace=True)
population1 = pop1_iloc

population1
```

#### 2) 생활인구

```python
df2 = pd.read_csv("data\서울시 우리마을가게 상권분석서비스(상권-생활인구).csv", encoding="cp949") 

pop2_2022 = df2.loc[(df2['기준 년코드'] == 2022) & (df2['기준_분기_코드'] == 4)].iloc[:,[4,6,9,10,11,12,13,14]]

pop2_2022['ave_age2'] = (15*pop2_2022.iloc[:,2]+25*pop2_2022.iloc[:,3]+35*pop2_2022.iloc[:,4]+45*pop2_2022.iloc[:,5]+55*pop2_2022.iloc[:,6]+65*pop2_2022.iloc[:,7])/pop2_2022.iloc[:,1]
pop2_2022['sum_age2'] = 15*pop2_2022.iloc[:,2]+25*pop2_2022.iloc[:,3]+35*pop2_2022.iloc[:,4]+45*pop2_2022.iloc[:,5]+55*pop2_2022.iloc[:,6]+65*pop2_2022.iloc[:,7]

pop2_iloc = pop2_2022.iloc[:,[0,1,-2,-1]]
population2 = pop2_iloc

population2
```

#### 3) 직장인구

```python
df3 = pd.read_csv("data\서울시 우리마을가게 상권분석서비스(상권-직장인구).csv", encoding="cp949") 

pop3_2022 = df3.loc[(df3['기준_년월_코드'] == 2022) & (df3['기준_분기_코드'] == 4)].iloc[:,[4, 6, 9, 10, 11, 12, 13, 14]]

pop3_2022['ave_age3'] = (15*pop3_2022.iloc[:,2]+25*pop3_2022.iloc[:,3]+35*pop3_2022.iloc[:,4]+45*pop3_2022.iloc[:,5]+55*pop3_2022.iloc[:,6]+65*pop3_2022.iloc[:,7])/pop3_2022.iloc[:,1]
pop3_2022['sum_age3'] = 15*pop3_2022.iloc[:,2]+25*pop3_2022.iloc[:,3]+35*pop3_2022.iloc[:,4]+45*pop3_2022.iloc[:,5]+55*pop3_2022.iloc[:,6]+65*pop3_2022.iloc[:,7]

pop3_iloc = pop3_2022.iloc[:,[0,1,-2,-1]]
population3 = pop3_iloc

population3
```

#### 4) 점포


```python
df = pd.read_csv("data\서울시 우리마을가게 상권분석서비스(상권-점포).csv", encoding="cp949")

store_2022 = df.loc[(df['기준_년_코드'] == 2022) & (df['기준_분기_코드'] == 4)]
store_iloc = store_2022.iloc[:,[4, 7, 8]]


target_names = ['편의점', '슈퍼마켓', '커피-음료', '분식전문점', '패스트푸드점', 
                '제과점', '호프-간이주점', '치킨전문점' ]
target_rows = store_iloc[store_iloc['서비스_업종_코드_명'].isin(target_names)]

data = {
    '상권_코드' : [],
    '점포_수' : []
}

for code in target_rows['상권_코드'].unique():
    temp = target_rows[target_rows['상권_코드'] == code]
    sum_of_temp = temp['점포_수'].sum(skipna=True)

    data['상권_코드'].append(code)
    data['점포_수'].append(sum_of_temp)

store = pd.DataFrame(data)
store
```

#### 5) 아파트

```python
df = pd.read_csv("data\서울시 우리마을가게 상권분석서비스(상권-아파트).csv", encoding="cp949") 

apt_2022 = df.loc[(df['기준_년_코드'] == 2022) & (df['기준_분기_코드'] == 4)]
apt_iloc = pd.DataFrame(apt_2022.iloc[:,[4, 6]])
apt_iloc.drop_duplicates(subset=['상권_코드'], keep='first', inplace=True)

apt = apt_iloc
apt
```

### 2. 최종 데이터프레임

```python
from functools import reduce

dfs = [population1, population2, population3, store, apt]
DF = reduce(lambda left, right: pd.merge(left, right, on='상권_코드', how='inner'), dfs)
```

```python
DF.drop_duplicates(subset=['상권_코드'], keep='first', inplace=True, ignore_index = True)

DF['총_인구수'] = DF.iloc[:,1] + DF.iloc[:,4] + DF.iloc[:,7]
DF['평균나이'] = (DF.iloc[:,3] + DF.iloc[:,6] + DF.iloc[:,9])/DF.loc[:,'총_인구수']

display(DF.head())
```

```python
DataFrame = DF.iloc[:,[0,12,13,10,11]]
DataFrame
```

### 3. X 추출 및 scaling

```python
import matplotlib.pyplot as plt

X = DataFrame.iloc[:, 1:]
plt.boxplot(X)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

plt.boxplot(X_scaled)
```

## II. clustering

### 1. 최적의 클래스 개수 찾기

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
sil_scores = []

for n_clusters in range(2,20):
    kmeans = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, cluster_labels))
```

```python
plt.figure(figsize = (20,10))

plt.subplot(1,2,1)
plt.plot(range(2,20), inertia, marker='o', lw=2, ms=10)
plt.xlabel("Number of clusters")
plt.xticks(range(2,20,1))
plt.title("K-Means Inertia")

plt.subplot(1,2,2)
plt.bar(range(2,20), sil_scores, )
plt.xlabel("Number of clusters")
plt.xticks(range(2,20,1))
plt.title("Silhouette Score")

plt.show()
```

### 2. clustering: n_clusters = 4

```python
n_clusters = 4

kmeans = KMeans(n_clusters= n_clusters, random_state = 10)
kmeans.fit(X_scaled)

print(kmeans.labels_)
```
